```python
import pandas as pd
import os

# Folder containing the CSV files
folder_path = 

column_names = ['fold', 'epoch', 'train_log_loss', 'test_log_loss', 'train_auc_roc', 'test_auc_roc', 'train_auc_pr', 'test_auc_pr']

rowCondition = "last" #get "last" row or "best" test set AUC-ROC

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('_results.csv')]

# Group files by model name
model_files = {}
for file in csv_files:
    model_name = file.split('_fold')[0]
    if model_name not in model_files:
        model_files[model_name] = []
    model_files[model_name].append(file)

# Iterate through each model and calculate the average of the last rows
averages = {}
std_devs = {}
for model_name, files in model_files.items():
    last_rows = []
    for file in files:
        df = pd.read_csv(os.path.join(folder_path, file), names=column_names)
        if rowCondition == "last":
            last_row = df.iloc[-1]
            if last_row['epoch'] != 49:
                last_row = df.iloc[-2]
        elif rowCondition == "best":
            last_row = df.loc[df['test_auc_roc'].idxmax()]
        last_rows.append(last_row)

    # Calculate the average and standard deviation of the last rows
    avg_last_row = pd.concat(last_rows, axis=1).drop('fold', axis=0).mean(axis=1)
    std_dev_last_row = pd.concat(last_rows, axis=1).drop('fold', axis=0).std(axis=1)
    averages[model_name] = avg_last_row
    std_devs[model_name] = std_dev_last_row


# Print the averages and standard deviations for each model
for model_name, avg in averages.items():
    std_dev = std_devs[model_name]
    avg_auc_roc = avg['test_auc_roc']
    avg_auc_pr = avg['test_auc_pr']
    std_auc_roc = std_dev['test_auc_roc']
    std_auc_pr = std_dev['test_auc_pr']

    print(f"Model: {model_name}")
    print(f"Average AUC-ROC: {avg_auc_roc:.3f} ± {std_auc_roc:.3f}")
    print(f"Average AUC-PR: {avg_auc_pr:.3f} ± {std_auc_pr:.3f}")
```
